##Step 1: Split of the Kaggle traning dataset on training and validation I datasets

```python
# Imports
import pandas as pd
from sklearn.model_selection import ShuffleSplit
```

```python
# Settings
dataTrainPath = 'data\\data_original\\train.csv'
dataTestPath = 'data\\data_original\\test.csv'
newDataTrainPath = 'data\\data_derived\\train.csv'
newDataTestPath = 'data\\data_derived\\test.csv'
newDataTrain1Path = 'data\\data_derived\\train1.csv'
newDataTrain2Path = 'data\\data_derived\\train2.csv'
```

```python
dataTrain = pd.read_csv(dataTrainPath)
dataTest = pd.read_csv(dataTestPath)
```

```python
dataTrain.head(5)
```

```python
dataTest.head(5)
```

```python
def survRate(data):
    survIdx = data.loc[:,'Survived'] == 1
    survRate = survIdx.sum() / survIdx.size
    return survRate
```

```python
print('Data Train size (Pos Rate) = %d (%.2f)' % (dataTrain.shape[0], survRate(dataTrain)))
print('Data Test size = ', dataTest.shape[0])
```

```python
dataSplit = next(ShuffleSplit(1, 0.3).split(dataTrain.index.values))
dataTrain1Idx = dataSplit[0]
dataTrain2Idx = dataSplit[1]
dataTrain1 = dataTrain.loc[dataTrain1Idx, :]
dataTrain2 = dataTrain.loc[dataTrain2Idx, :]
print('Data Train 1 size (Pos Rate) = %d (%.2f)' % (dataTrain1.shape[0], survRate(dataTrain1)))
print('Data Train 2 size (Pos Rate) = %d (%.2f)' % (dataTrain2.shape[0], survRate(dataTrain2)))
```

```python
dataTrain.to_csv(newDataTrainPath, index=False)
dataTrain1.to_csv(newDataTrain1Path, index=False)
dataTrain2.to_csv(newDataTrain2Path, index=False)
dataTest.to_csv(newDataTestPath, index=False)
```
