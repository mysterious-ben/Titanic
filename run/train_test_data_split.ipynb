## Training-validation data split
- **Step 0: Training-validation data split**
	- Trainig data set is used to fit classifiers (70% of the Kaggle training data set)
	- Validation I data set is used to estimate OOS performance (30% of the Kaggle training data set)
	- Validation II data set is the data set for the Kaggle submission (Kaggle test data set)
- Step 1: Data analysis
- Step 2: Classification
- Step 3: Model Validation

```python
# Imports
import pandas as pd
from sklearn.model_selection import ShuffleSplit
from IPython.display import display
```

```python
# Settings
dataTrainPath = 'data\\data_original\\train.csv'
dataTestPath = 'data\\data_original\\test.csv'
newDataTrainPath = 'data\\data_derived\\train.csv'
newDataTestPath = 'data\\data_derived\\test.csv'
newDataTrain1Path = 'data\\data_derived\\train1.csv'
newDataTrain2Path = 'data\\data_derived\\train2.csv'
```

```python
dataTrain = pd.read_csv(dataTrainPath)
dataTest = pd.read_csv(dataTestPath)
display(dataTrain.head(5))
display(dataTest.head(5))
```

```python
def survRate(data):
    survIdx = data.loc[:,'Survived'] == 1
    survRate = survIdx.sum() / survIdx.size
    return survRate
```

```python
print('Kaggle training set size (Pos Rate) = %d (%.2f)' % (dataTrain.shape[0], survRate(dataTrain)))
dataSplit = next(ShuffleSplit(1, 0.3).split(dataTrain.index.values))
dataTrain1Idx = dataSplit[0]
dataTrain2Idx = dataSplit[1]
dataTrain1 = dataTrain.loc[dataTrain1Idx, :]
dataTrain2 = dataTrain.loc[dataTrain2Idx, :]
print('Training set size (Pos Rate) = %d (%.2f)' % (dataTrain1.shape[0], survRate(dataTrain1)))
print('Validation I set size (Pos Rate) = %d (%.2f)' % (dataTrain2.shape[0], survRate(dataTrain2)))
print('Validation II (= Kaggle test) set size = ', dataTest.shape[0])
```

```python
dataTrain.to_csv(newDataTrainPath, index=False)
dataTrain1.to_csv(newDataTrain1Path, index=False)
dataTrain2.to_csv(newDataTrain2Path, index=False)
dataTest.to_csv(newDataTestPath, index=False)
```
